```python
%matplotlib widget

import matplotlib.pyplot as plt
import numpy as np

# Dados para o gráfico
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Criar o gráfico
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.title('Gráfico Interativo com %matplotlib widget')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

# O gráfico será exibido automaticamente
```

```python

x = np.linspace(0, 2 * np.pi, 200)
y = np.sin(x)

fig, ax= plt.subplots()
ax.plot(x,y)
plt.show()
```
