Com o método <code> .query() </code> do Pandas permite filtrar dados de um DataFrame utilizando uma <b> expressão </b> de consulta semelhante à linguagem SQL. Permitindo selecionar linhas com base em condições específicas definidas na expressão de consulta, facilitando a filtragem dos dados. 

Ela aceita um text (string) contendo a expressão de consulta e retorna uma df contendo apenas as linhas que atendem às condições especificadas. 


Pode incluir operadores lógicos, como <b> and, or e not</b> bem como operadores de comparação <b> ==, <>, <=> </b>

```python
import pandas as pd

vendas = pd.read_csv("practice\serenatto_2sem_2023.csv")
vendas.head()
```

```python
## Query para filtrar dados em duas ou mais colunas

vendas.query('valor > 10 and metodo_pagamento != "Dinheiro" ')

## Nesse exemplo, estou filtrando o df para selecionar apenas as linhas em que o valor
## do produto é maior que 10(R$10) e o método do pagamento é qualquer um, exceto "Dinheiro".
```

#### Query com o operador @

Permitindo referenciar variáveis externas dentro da expressão de consulta usando o operador <b> @ </b>. Sendo útil para criar consultas dinâmicas com base em variáveis definidas previamente. 

```python
produtos = ['Café au lait', 'Espresso', 'Cappuccino']

vendas.query('produto in @produtos and metodo_pagamento == "PIX" ')
```

Nesse exemplo, estou definindo o df para selecioanr apenas as linhas em que tenham os produtos passados pela lista <code> produtos </code> e que o método de pagamento seja em PIX.
