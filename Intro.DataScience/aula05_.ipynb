```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

tmdb = pd.read_csv("tmdb_5000_movies.csv")
tmdb.head()
```

```python
total_de_outros_filmes_por_lingua = tmdb.query("original_language != 'en'")["original_language"].value_counts()
total_de_outros_filmes_por_lingua.head()
```

```python
plt.figure(figsize=(16,8)) ## Utilizando o plt pyplot para definir o tamanho do meu gráfico.

sns.countplot(data=tmdb.query("original_language != 'en'"),
              order=total_de_outros_filmes_por_lingua.index, ## Ordem dos dados apresentados baseado na minha variável
              palette="mako", ## Definindo uma paleta de cor - escolhi essa através da documentação do seaborn
              hue="original_language", ## Entrada para plotagem de dados no formato, nesse caso o 'original_language'
              hue_order=total_de_outros_filmes_por_lingua.index, ## Definindo a ordem das cores baseado no index.
              stat="percent", ## Exibir em formato de percentual %
              x="original_language")
plt.title("Distribuição da língua original nos filmes exceto em Inglês")
plt.show()
```
