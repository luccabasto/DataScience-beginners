Fazendo um review geral do que aprendi até aqui, com o intuito de reforçar. 

Vamos explorar algumas de suas possibilidades. Utilizando a basedo TMDB, filtroremos os filmes nas línguas francesa, espanhola, e alemã, com valores vlálidos de receita (revenue) e orçamento (budget). Em sequência, criando 3 gráficos de dispersão com retas de regressão linear .lmplot() para cada língua.

```python
import pandas as pd
import seaborn as sns

# Importando a base de dados
tmdb = pd.read_csv("https://raw.githubusercontent.com/alura-cursos/data-science-analise-exploratoria/main/Aula_0/tmdb_5000_movies.csv")

# Filtrando os dados para as colunas que desejo

dados = tmdb[["title", "original_language", "budget", "revenue"]]
linguas = ["fr", "es", "de"] 
```

```python
# Mantendo dados com valores de receita e orçamento válidos e nas línguas (fr, es, de)

filtro_linguas_desejada_com_revenue_budget = dados.query("revenue > 0 and budget > 0 and original_language == @linguas")
```

```python
# Construindo o gráfico de dispersão mais a reta de regressão linear para cada caso

sns.lmplot(data=filtro_linguas_desejada_com_revenue_budget, x="budget", y="revenue", col="original_language", hue="original_language", ci=None, height=4, scatter_kws={"s": 30, "alpha": 0.5})
```

Consigo notar diferenças entre as produções em cada uma das línguas, mostrando as tendências da relação entre o orçamento utilizado e a receita gerada pelos filmes. Seria necessário um número maior de filmes para cada uma delas para gerarmos melhores interpretações com os dados.
